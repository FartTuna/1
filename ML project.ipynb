## –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –æ—Ç—Ç–æ–∫–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤

–†–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ, –Ω–∞ –æ–±—É—á–∞—é—â–∏–µ –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã–µ –≤—ã–±–æ—Ä–∫–∏ –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∏—Ä—É–µ–º –∏—Ö, —Ç.–∫. –Ω–µ –≤—Å–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –≤ –Ω–∞—à–µ–π –±–∞–∑–µ –∏–º–µ—é—Ç –±–∏–Ω–∞—Ä–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç:

<div class="alert alert-warning">

**–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞**

–ú–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä stratify = y –≤ —Ñ—É–Ω–∫—Ü–∏–∏ train_test_split( ), –≥–¥–µ y ‚Äì —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∞—è —Ü–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è. –î–∞–Ω–Ω—ã–µ —Ä–∞–∑–¥–µ–ª—è—Ç—Å—è –Ω–∞ –¥–≤–∞ –∫–ª–∞—Å—Ç–µ—Ä–∞ (–Ω–∞–±–ª—é–¥–µ–Ω–∏—è –∫–ª–∞—Å—Å–∞ 0 –∏ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è –∫–ª–∞—Å—Å–∞ 1), –∏ –≤ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫—É –ø–æ–ø–∞–¥—É—Ç –ø–æ 20% –Ω–∞–±–ª—é–¥–µ–Ω–∏–π –∏–∑ –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Ç–µ—Ä–∞:


    
–¢–æ –µ—Å—Ç—å, –µ—Å–ª–∏ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –≤ –æ—Ä–∏–≥–∏–Ω–∞–ª—å–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –±—ã–ª–æ 2:5, —Ç–æ –≤ –æ–±—É—á–∞—é—â–µ–π –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∞—Ö –æ–Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—Å—è —Ç–∞–∫–∏–º –∂–µ. –¢–∞–∫–æ–π –ø–æ–¥—Ö–æ–¥ –≤ —á–µ–º-—Ç–æ —Ä–µ—à–∞–µ—Ç –ø—Ä–æ–±–ª–µ–º—É –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤.

–§—É–Ω–∫—Ü–∏—è train_test_split( ) —Å—Ç—Ä–æ–µ–Ω–∞ —Ç–∞–∫, —á—Ç–æ –º—ã –Ω–µ —Å—Ç–∞–ª–∫–∏–≤–∞–µ–º—Å—è —Å –ø–æ–¥–æ–±–Ω–æ–π –ø—Ä–æ–±–ª–µ–º–æ–π. –ù–æ —è –≤—Å–µ –∂–µ –ø–∏—à—É –ø—Ä–æ —ç—Ç–æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä, —á—Ç–æ–±—ã –æ–±—Ä–∞—Ç–∏—Ç—å –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ –ø—Ä–æ–±–ª–µ–º—É –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞ :)</div>

<div class="alert alert-info">

    
**–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞**
    
–û–≥—Ä–æ–º–Ω–æ–µ —Å–ø–∞—Å–∏–±–æ –∑–∞ —ç—Ç–æ—Ç —Å–æ–≤–µ—Ç! –ö–æ–≥–¥–∞ —Ä–µ–≤—å—é–µ—Ä—ã –∏–º–µ—é—Ç –¢–ê–ö–û–ô –ø–æ–¥—Ö–æ–¥ –∫ –ø—Ä–æ–≤–µ—Ä–∫–µ, —è –≤–µ—Ä—é, —á—Ç–æ –æ–Ω–∏ –∏—Å–∫—Ä–µ–Ω–Ω–µ –ª—é–±—è—Ç –¥–µ–ª–æ, –∫–æ—Ç–æ—Ä—ã–º –∑–∞–Ω–∏–º–∞—é—Ç—Å—è :)
</div>

<div class="alert alert-success">

**–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–≤–µ—Ä–∞ v2**

–°–ø–∞—Å–∏–±–æ :)

```python
X = data.drop('Churn', axis = 1)
y = data['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)

scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)
X_test_st = scaler.transform(X_test)
```

–°–ø–∏—Å–æ–∫ –æ–±—É—á–∞—é—â–∏—Ö –º–æ–¥–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –º—ã –ø–ª–∞–Ω–∏—Ä—É–µ–º –ø—Ä–∏–º–µ–Ω–∏—Ç—å –∏ —Ñ—É–Ω–∫—Ü–∏—è, –æ–±—É—á–∞—é—â–∞—è –º–æ–¥–µ–ª—å –∏ –æ—Ç–æ–±—Ä–∞–∂–∞—é—â–∞—è –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏ –ø–æ —Ç—Ä—ë–º –º–µ—Ç—Ä–∏–∫–∞–º:

<div class="alert alert-danger">

**–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–≤–µ—Ä–∞**

–ù–µ–æ–±—Ö–æ–¥–∏–º–æ —É–∫–∞–∑–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä random_state –≤ —Ñ—É–Ω–∫—Ü–∏–∏ RandomForestClassifier( ), –∏–Ω–∞—á–µ –ø—Ä–∏ –ø–µ—Ä–µ–∑–∞–ø—É—Å–∫–µ –∫–æ–¥–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –±—É–¥—É—Ç —Ä–∞–∑–Ω—ã–µ. –ü–æ–ø—Ä–∞–≤—å, –ø–æ–∂–∞–ª—É–π—Å—Ç–∞, –∏ –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–π –∏–∑–º–µ–Ω–µ–Ω–∏—è :)</div>
    
    
<div class="alert alert-info">

    
**–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞**
    
–ò—Å–ø—Ä–∞–≤–∏–ª!
</div>

<div class="alert alert-success">

**–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–≤–µ—Ä–∞ v2**

–¢–µ–ø–µ—Ä—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–µ –ø–æ—Ç–µ—Ä—è–µ—Ç—Å—è üëçüèª

```python
models = [
    LogisticRegression(),
    RandomForestClassifier(random_state = 0),
]

def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_test, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_test, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_test, y_pred))) 
```
